```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("/content/ORDERS.csv", encoding="ISO-8859-1")

# Aggregate sales by region and province
sales_by_region = df.groupby(["REGION", "PROVINCE"])["SALES"].sum().reset_index()

# Plot sales by region
plt.figure(figsize=(12, 6))
sns.barplot(data=sales_by_region, x="REGION", y="SALES", estimator=sum, ci=None)
plt.xticks(rotation=45)
plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.show()
```
